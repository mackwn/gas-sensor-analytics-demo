# Helper Functions

### Helper Function

```python
# Take a regressor as input and compare how score vs number of samples
def gen_learning_curve(regressor,X,y,X_cv,y_cv):
    n_arr = list(range(50,250,25)) + list(range(250,len(X),500))
    scores_learn = []
    scores_learn_cv = []
    for n in n_arr:
        reg = regressor.fit(X[0:n,:], y[0:n])
        scores_learn.append(reg.score(X[0:n,:],y[0:n]))
        scores_learn_cv.append(reg.score(X_cv,y_cv))
    return n_arr, scores_learn, scores_learn_cv

```

```python
def goodness_of_fit_plot(y, y_pred, y_cv, y_pred_cv, error_per, title_text):
    fig, ax = plt.subplots(1,3,figsize=(16,5))

    fig.suptitle('Goodness of Fit - {}'.format(title_text), fontsize=20)

    ax[0].scatter(y_cv,y_cv-y_pred_cv, label='CV')
    ax[0].scatter(y,y-y_pred, label='Training')
    ax[0].set_title('Y value vs Residual')
    ax[0].set_ylabel('Actual Y value')
    ax[0].set_xlabel('Residuals')
    ax[0].legend()

    ax[1].hist(y_cv-y_pred_cv, density=True, range=(-40,40), bins=40, label='CV')
    ax[1].set_title('Hist. of Residuals - CV')
    ax[1].set_xlabel('Residuals')
    ax[1].legend()

    ax[2].hist(error_per,range=(0,1),cumulative=True, density=True, bins=40, label='CV')
    ax[2].set_ylim(.6)
    ax[2].set_title('Hist.of Abs. Perc. Err - CV')
    ax[2].set_xlabel('abs(y_pred-y)/y')
    ax[2].legend()

    plt.tight_layout()
```
